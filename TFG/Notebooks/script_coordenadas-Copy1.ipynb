Esto lo hago porque en mi caso los números venían muy raros, los decimales los separaba por '.' y entonces no los detectaba como num, por tanto cambié . por , y convierto en float

```python
accidentes['coordenada_x_local'] = accidentes['coordenada_x_local'].str.replace(',', '.').astype(float)
accidentes['coordenada_y_local'] = accidentes['coordenada_y_local'].str.replace(',', '.').astype(float)
```

Con este for aplico convierto a lat lon y lo meto en una variable "coordenadas", el resultado del código es un vector lat, lon como el de arriba

```python
accidentes['coordenadas'] = ''
for i in range(len(accidentes)):
    if (accidentes['coordenada_x_local'][i] > 0) or (accidentes['coordenada_x_local'][i] > 0): 
        accidentes['coordenadas'][i] = utm.to_latlon(accidentes['coordenada_x_local'][i], accidentes['coordenada_y_local'][i], 30, northern = True)
    else:
        accidentes.drop(i, inplace = True)
```
